```python
cat_cols = ['AP006','MB006']
def encode(encoder, x):
    len_encoder = len(encoder)
    try:
        ind = encoder[x]
    except KeyError:
        ind = len_encoder
    return ind

encoders = [{} for cat in cat_cols]

for i, cat in enumerate(cat_cols):
    print('encoding %s'%cat, end='\n')
    encoders[i] = {l:ind for ind, l in enumerate(df[cat].unique())}
    df[cat] = df[cat].apply(lambda x:encode(encoders[i],x))
    
embed_sizes = [len(encoder) + 1 for encoder in encoders]
```
